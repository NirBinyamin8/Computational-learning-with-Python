```python
# train a linear regression model to predict medical expenses (the last column) from other features

# perform all your computations (solve the task) before the questions part, in a clear and effective manner
# in the questions part only print answers based on your solution
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import pearsonr


def smokerfunc(smoker):
    return 1 if smoker=='yes'else  0
def isFemale(word):
    if word == "female" : 
        return 1
    else:
        return 0

def isMale(word):
    if word == "male":
        return 1
    else:
        return 0
    
def isSouthwest(word):
    if word == "southwest":
        return 1
    else:
        return 0
    
def isNorthwest(word):
    if word == "northwest":
        return 1
    else:
        return 0
    
def isNorthEast(word):
    if word == "northeast":
        return 1
    else:
        return 0

    
def isSouthEast(word):
    if word == "southeast":
        return 1
    else:
        return 0


df = pd.read_csv('insurance_data.csv')
df['smoker\yes']=df['smoker'].apply(smokerfunc)
df['male'] = df['gender'].apply(isMale)
df['female']=df['gender'].apply(isFemale)
df['SoutWest'] = df['region'].apply(isSouthwest)
df['Northwest'] = df['region'].apply(isNorthwest)
df['NorthEast'] = df['region'].apply(isNorthEast)
df['SouthEast'] = df['region'].apply(isSouthEast)
df = df.sample(frac=1).reset_index(drop=True)
df=df[['age','bmi','children','smoker\yes','male','female','SoutWest','Northwest','NorthEast','SouthEast', 'expenses']]
df_train = df.iloc[:1000]
df_test = df.iloc[1000:]


y=df_train['expenses']
x=df_train[['age','bmi','children','smoker\yes','male','female','SoutWest','Northwest','NorthEast','SouthEast']]
x=sm.add_constant(x)
model=sm.OLS(y,x)
result=model.fit()


x_test = df_test[['age','bmi','children','smoker\yes','male','female','SoutWest','Northwest','NorthEast','SouthEast']]
y_test = df_test['expenses']

x_test = sm.add_constant(x_test)
y_predict = result.predict(x_test)

display(pearsonr(y_predict, y_test)[0])
result.summary()
```
